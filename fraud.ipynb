```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                             f1_score, matthews_corrcoef, confusion_matrix)
from imblearn.over_sampling import SMOTE
import joblib
```

```python
# Load CSV
data = pd.read_csv("data/creditcard.csv")
data.head(5)
```

```python
# Dataset description
print("Dataset description:\n", data.describe())

# Count Fraud vs Normal
fraud = data[data['Class'] == 1]
valid = data[data['Class'] == 0]

print("Fraud Cases:", len(fraud))
print("Valid Transactions:", len(valid))
print("Outlier Fraction:", len(fraud)/float(len(valid)))
```

```python
# Correlation Heatmap
corrmat = data.corr()
plt.figure(figsize=(12, 9))
sns.heatmap(corrmat, vmax=0.8, square=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Drop 'Class' and 'Time' (Time not useful for prediction)
X = data.drop(['Class', 'Time'], axis=1)
Y = data['Class']

# Handle imbalanced data using SMOTE
smote = SMOTE(random_state=42)
xData, yData = smote.fit_resample(X, Y)

print("After SMOTE, counts of label '1':", sum(yData==1))
print("After SMOTE, counts of label '0':", sum(yData==0))
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(
    xData, yData, test_size=0.2, random_state=42
)
```

```python
rfc = RandomForestClassifier(class_weight='balanced', random_state=42)
rfc.fit(xTrain, yTrain)
```

```python
#evaluate model
yPred = rfc.predict(xTest)

acc = accuracy_score(yTest, yPred)
prec = precision_score(yTest, yPred)
rec = recall_score(yTest, yPred)
f1 = f1_score(yTest, yPred)
MCC = matthews_corrcoef(yTest, yPred)

print("Random Forest Classifier Metrics:")
print("Accuracy:", acc)
print("Precision:", prec)
print("Recall:", rec)
print("F1-Score:", f1)
print("Matthews Corr Coef:", MCC)

# Confusion matrix
LABELS = ['Normal', 'Fraud']
conf_matrix = confusion_matrix(yTest, yPred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, xticklabels=LABELS, yticklabels=LABELS, annot=True, fmt="d")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
joblib.dump(rfc, "model.pkl")
print("Model saved as model.pkl")
```

```python
def predict_fraud(input_data):
    """
    input_data: 2D array or dataframe with 29 features (V1-V28 + Amount)
    """
    prediction = rfc.predict(input_data)
    prediction_proba = rfc.predict_proba(input_data)[0][1]  # Fraud probability
    return {
        "Prediction": "Fraud" if prediction[0]==1 else "Normal",
        "Fraud Probability": round(prediction_proba*100, 2)
    }
```

```python
# Sample transaction
sample_input = pd.DataFrame([{
    'V1': 0.01, 'V2': 4.14, 'V3': -6.24, 'V4': 6.68,
    'V5': 0, 'V6': 0, 'V7': 0, 'V8': 0, 'V9': 0, 'V10': 0,
    'V11': 0, 'V12': 0, 'V13': 0, 'V14': 0, 'V15': 0,
    'V16': 0, 'V17': 0, 'V18': 0, 'V19': 0, 'V20': 0,
    'V21': 0, 'V22': 0, 'V23': 0, 'V24': 0, 'V25': 0,
    'V26': 0, 'V27': 0, 'V28': 0, 'Amount': 1.0
}])

predict_fraud(sample_input)
```
